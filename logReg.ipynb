```python
#import mysql.connector
#import mysql
import pandas as pd
import numpy as np
from joblib import dump
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import  LogisticRegression
from sklearn.metrics import precision_recall_fscore_support as score



df1.head()
```

```python
df1=pd.read_csv("titanic_csv_1.csv",sep=";")
df2=pd.read_csv("titanic_db_1.csv",sep=";")
df3=pd.read_csv("titanic_db_2.csv",sep=";")
```

```python
df2.head()
```

```python
df1.drop_duplicates(subset='PassengerId',inplace=True)
df2.drop_duplicates(subset='PassengerId',inplace=True)
df3.drop_duplicates(subset='PassengerId',inplace=True)
df=pd.merge(df1,df2,on='PassengerId',how='outer')
df=df.append(df3)
#df = pd.concat([df,df3]) #ici on a pris en ligne par defaut
#trouver les colone qui ont des nan
df.isna().sum()
```

```python
#df.drop(['SibSp','Parch','Ticket','Cabin','Embarked'], axis= 1,inplace =True)
df.shape
```

```python
df.isna().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df.isna().sum()
```

```python
df.head()
```

```python
if 'Mr.' in 'Hirvonen, Mrs. Alexander (Helga E Lindqvist)':
    print('male')
else :
    print('female')
```

```python
#quand on fait df.apply(lambda x:) notre df est contenue dans le x ainsi donc chaque appel qu'on voudra faire de notre dataframe 
df['sex2']=df.apply(lambda x:'male' if 'Mr.' in x['Name'] else 'female', axis=1)
```

```python
df.head()
```

```python
df['Sex']=df.sex2
df.drop(['sex2'], axis= 1,inplace =True)
df.head()
df.isna().sum()
```

```python
df.head(30)
```

```python
df.groupby('Pclass')['Fare'].describe()
#[0,14][14,26][26,+]
```

```python
#[0,14][14,26][26,+]
def classification(fare):
    if fare>=0 and fare<14:
         return 3
    elif fare>=14 and fare<26:
        return 2
    elif fare>26:
        return 1
    else:
        return 0
        

#df['Pclass'].fillna(classification(df.Pclass))
df['Pclass'].astype(float).isna()
```

```python
classification(20)
df['Pclass']=df.apply(lambda x:classification(x['Fare']) if np.isnan(x['Pclass']) else x['Pclass'], axis=1)
```

```python
df.isna().sum()
```

```python
def codeSex(Sex):
    if Sex=='male':
         return 1
    else:
        return 0
    
df['SexCode']=df.apply(lambda x:codeSex(x['Sex']), axis=1)
df.head()
```

```python
#Tout ce qui ont survuved not null
dfsurvived = df[df['Survived'].notnull()]
dfsurvived.isna().sum()
```

```python
y=dfsurvived['Survived']
x=dfsurvived[['Age','SexCode','Pclass']]
```

```python
dfsurvived.dtypes
```

```python
dfsurvived.head()
```

```python


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
log_regression_alg = LogisticRegression()
log_regression_alg.fit(x_train, y_train)
test_predictions = log_regression_alg.predict(x_test)
precision, recall, _, _ = score(y_test, test_predictions, pos_label=1, average='binary')
accuracy = round((test_predictions==y_test).sum() / len(test_predictions),3)
#Prcision = Vrais positifs / (Vrais positifs + Faux positifs)
#recall = vrais positifs / (vrais positifs + faux ngatifs )
print('Precision: {} / Recall: {} / Accuracy: {}'.format(round(precision, 3), round(recall, 3), accuracy))
```
